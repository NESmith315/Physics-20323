```python
# write/copy code here
class Particle(object):
    def __init__(self, m, x, y, z, vx, vy, vz):
        self.mass = m
        self.x = x
        self.y = y
        self.z = z
        self.xVel = vx
        self.yVel = vy
        self.zVel = vz
    
    def ret_KE(self):
        return 0.5 * self.mass * (self.xVel ** 2 + self.yVel ** 2 + self.zVel ** 2)
    
    def dist_betw(self, other):
        return np.sqrt((self.x - other.x)**2 + (self.y - other.y)**2 + (self.z - other.z)**2)
    
    def update_pos(self, dt):
        self.x = self.x + self.xVel * dt
        self.y = self.y + self.yVel * dt
        self.z = self.z + self.zVel * dt

class ChargedParticle(Particle):
    def __init__(self, m, x, y, z, vx, vy, vz, c):
        super().__init__(m, x, y, z, vx, vy, vz)
        self.charge = c
    
    def simple_interaction(self, other):
        if self.dist_betw(other) < 0.25:
            print("interaction")
            if np.sign(self.charge) != np.sign(other.charge):
                self.xVel = 0
                self.yVel = 0
                self.zVel = 0
                self.charge = 0
                other.xVel = 0
                other.yVel = 0
                other.zVel = 0
                other.charge = 0
                print("merge")
            else:
                self.xVel *= -1 * (self.charge + other.charge) * (self.mass/other.mass)
                self.yVel *= -1 * (self.charge + other.charge) * (self.mass/other.mass)
                self.zVel *= -1 * (self.charge + other.charge) * (self.mass/other.mass)
                other.xVel *= -1 * (self.charge + other.charge) * (other.mass/self.mass)
                other.yVel *= -1 * (self.charge + other.charge) * (other.mass/self.mass)
                other.zVel *= -1 * (self.charge + other.charge) * (other.mass/self.mass)
                print("repel")
```
